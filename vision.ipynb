```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np

from google.colab.patches import cv2_imshow
import cv2
import numpy as np

# Load the input image
input_image = cv2.imread('/content/blue circle.png')  # Adjust the file path accordingly

if input_image is None:
    print("Error: Unable to load the image. Please check the file path.")
else:
    # Function to make the blue circle white and white background black
    def process_image(image):
        # Convert the image to HSV format for better color manipulation
        hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

        # Define color range for the blue color
        lower_blue = np.array([69, 153, 217])
        upper_blue = np.array([69, 153, 217])

        # Create a mask for the blue color
        mask = cv2.inRange(hsv_image, lower_blue, upper_blue)

        # Create a mask for the white background
        white_background_mask = cv2.inRange(image, (200, 200, 200), (255, 255, 255))

        # Combine the blue circle mask and white background mask
        final_mask = cv2.bitwise_or(mask, white_background_mask)

        # Make blue white and white background black
        result_image = cv2.bitwise_and(image, image, mask=final_mask)
        result_image = cv2.bitwise_not(result_image)

        return result_image

    # Display the original image
    cv2_imshow(input_image)

    # Display the result with the blue circle made white and the white background black
    processed_image = process_image(input_image)
    cv2_imshow(processed_image)
```
